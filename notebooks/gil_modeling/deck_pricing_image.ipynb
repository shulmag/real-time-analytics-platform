```python
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
import re
import time

browser = webdriver.Chrome()

# Read CSV
data = pd.read_csv('/Users/gil/git/ficc/notebooks/gil_modeling/finra.csv')
names = data.iloc[:, 0].tolist()

# List to store names that have relevant keywords in their content
names_with_relevant_keywords = []

# List to store names that have "Municipal Bonds" in their content
names_with_mbonds = []

# Keywords to look for in the page source
keywords = ["muni", "fixed income", "capital market", "municipal", "sma", "bond"]

for name in names:
    # Google the name
    browser.get('https://www.google.com/')
    search_box = browser.find_element(by=By.NAME, value='q')
    search_box.send_keys(name)
    search_box.submit()
    time.sleep(2)  # Wait for page load

    # Click the first link
    try:
        first_link = browser.find_element(by=By.CSS_SELECTOR, value='h3')
        first_link.click()
        time.sleep(3)  # Wait for page load
        
        # Check if any of the keywords exist in the page content
        if any(keyword.lower() in browser.page_source.lower() for keyword in keywords):
            names_with_relevant_keywords.append(name)
    except Exception as e:
        print(f"An error occurred with {name}. Error: {e}")

browser.quit()

# Save names_with_relevant_keywords to a CSV file
df = pd.DataFrame(names_with_relevant_keywords, columns=['Relevant Names'])
df.to_csv('relevant_names.csv', index=False)

print(names_with_relevant_keywords)








```
