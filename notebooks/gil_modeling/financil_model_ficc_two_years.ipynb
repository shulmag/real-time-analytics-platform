```python
import pandas as pd

# Constants and initial setup
quarters_corrected = ["Q2 2024", "Q3 2024", "Q4 2024", "Q1 2025", "Q2 2025", "Q3 2025", "Q4 2025", "Q1 2026"]
customer_growth_sequence = [1, 4, 10, 15, 22, 30, 40, 50]  # Customer growth per quarter
opex_growth_rate_annual = 0.15  # 15% annual growth rate for Operating Expenses
opex_growth_rate_quarterly = (1 + opex_growth_rate_annual) ** (1/4) - 1  # Quarterly growth rate
revenue_per_customer_increase_rate = 0.05  # 5% increase every 6 months
tax_rate = 0.21  # 21%
initial_revenue_per_customer = 25000
initial_operating_expenses = 331988  # Starting point for operating expenses

# DataFrame for financial projections
df_projections = pd.DataFrame(index=quarters_corrected, columns=["Customers", "Quarterly Revenue per Customer", "Revenue", "Operating Expenses", "Net Income"])

current_revenue_per_customer = initial_revenue_per_customer

for i, quarter in enumerate(quarters_corrected):
    # Update revenue per customer every 6 months
    if i % 2 == 1:  # Every 6 months (every two quarters starting from the second quarter in our list)
        current_revenue_per_customer *= (1 + revenue_per_customer_increase_rate)
    
    customers = customer_growth_sequence[i]
    revenue = current_revenue_per_customer * customers
    
    # Operating expenses growth
    if i != 0:  # Grow operating expenses starting from the second quarter
        initial_operating_expenses *= (1 + opex_growth_rate_quarterly)
    
    operating_income = revenue - initial_operating_expenses
    taxes = max(operating_income * tax_rate, 0)
    net_income = operating_income - taxes
    
    # Populate DataFrame
    df_projections.loc[quarter] = [customers, current_revenue_per_customer, revenue, initial_operating_expenses, net_income]

# Format the DataFrame for display
df_projections.style.format({
    "Quarterly Revenue per Customer": "${:,.2f}",
    "Revenue": "${:,.2f}",
    "Operating Expenses": "${:,.2f}",
    "Net Income": "${:,.2f}",
})
```
