```python
import redis
import pandas as pd
```

Check which yield curve coefficients exist for a specified `date`.

```python
date = '2024-05-15'
```

```python
exists, does_not_exist = [], []
for hour in range(9, 16):    # 9:30am to 3:59pm ET is when we get real time yield curve values
    for minute in range(60):
        if hour == 9 and minute < 30: continue    # ignore all times between 9am - 9:29am
        string_date = pd.to_datetime(f'{date} {hour}:{minute}').strftime('%Y-%m-%d:%H:%M')
        if yield_curve_redis_client.exists(string_date):
            exists.append(string_date)
        else:
            does_not_exist.append(string_date)
```
