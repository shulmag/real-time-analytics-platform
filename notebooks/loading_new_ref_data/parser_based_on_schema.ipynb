```python
# current_year, current_month, current_day = get_current_year_month_day()
dates = ['2024-07-01', '2024-07-02']    # each date must be in the YYYY-MM-DD format
```

```python
def get_xml_files_and_upload_to_bigquery(date: str):
    '''Almost identical to main.py::main().'''
    year, month, day = date.split('-')
    directory = f'{year}/{month}/{day}/'

    xml_files = get_xml_files_from_directory(directory)
    if len(xml_files) == 0:
        print(f'No XML files in directory: {directory}')    # warnings.warn(no_xml_files_message, RuntimeWarning)
    else:
        print(f'{len(xml_files)} present in directory: {directory}. Files are: {", ".join(xml_files[:3])} ... {", ".join(xml_files[-3:])}')
    
    upload_list_of_filenames_to_bigquery(xml_files)
    print(f'{len(xml_files)} files from {directory} uploaded to sp_nested BigQuery table')
```

```python
for date in dates:
    get_xml_files_and_upload_to_bigquery(date)
```
