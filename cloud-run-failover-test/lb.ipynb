```python
import requests
import time

def test_endpoint(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()  # Raise an exception for HTTP errors
        return response.json()
    except requests.RequestException as e:
        print(f"Request failed: {e}")
        return None

url = "http://34.54.53.62/"
failover_started = False
failover_completed = False
failover_start_time = None
failover_end_time = None
start_time = time.time()

while time.time() - start_time < 600:  # Run for 10 minutes
    response = test_endpoint(url)
    current_time = time.time()

    if response:
        region = response.get('region', 'Unknown')
        timestamp = response.get('timestamp', 'Unknown')
        unique_id = response.get('id', 'Unknown')

        print(f"Response received:")
        print(f"  Region: {region}")
        print(f"  Timestamp: {timestamp}")
        print(f"  ID: {unique_id}")

        if not failover_started:
            if region != 'west-failover':
                # We were expecting responses from 'west-failover', but got something else
                failover_started = True
                failover_start_time = current_time
                print(f"Failover started at {failover_start_time - start_time:.2f} seconds")

        if failover_started and not failover_completed:
            if region == 'east-primary':
                failover_end_time = current_time
                failover_completed = True
                failover_duration = failover_end_time - failover_start_time
                print(f"Failover completed at {failover_end_time - start_time:.2f} seconds")
                print(f"Total failover time: {failover_duration:.2f} seconds")
    else:
        print("No response received.")
        if not failover_started:
            failover_started = True
            failover_start_time = current_time
            print(f"Failover started at {failover_start_time - start_time:.2f} seconds")

    time.sleep(1)  # Wait for 1 second before next request

print("Test completed.")
```
