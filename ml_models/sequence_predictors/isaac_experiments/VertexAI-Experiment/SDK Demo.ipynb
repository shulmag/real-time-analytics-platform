The focus of this notebook is to demonstrate how to programatically configure and deploy a Vertex AI job. The main benefit of this is for larger jobs like backtesting model performance across several months where doing so on the GCP console directly is troublesome. In general, the SDK is easy to use - simply instantiate a CustomJob class, pass it a dictionary of relevant parameters and if doing a HyperparameterTuningJob, pass the CustomJob as an argument to the HyperparameterTuningJob class. Some variables, such as location, can be initiated just once when initiating aiplatform, and passing it as argument to individual jobs only serves to override the original value.

# VERTEX AI SDK

```python
import google.cloud.aiplatform as aip
from google.cloud.aiplatform import hyperparameter_tuning as hpt
from datetime import datetime
```

Define arguments for HPT job

```python
LOCATION = "us-central1"
STAGING_BUCKET = "gs://ficc-model-experiments"
PROJECT_ID ='eng-reactor-287421'
SERVICE_ACCOUNT = "964018767272-compute@developer.gserviceaccount.com"
JOB_NAME = "similar_trade_history_5yr_2LSTM_attention_FULL" #this is the display name on Vertex AI console
CONTAINER_URI = "us-east4-docker.pkg.dev/eng-reactor-287421/custom-train-job/model-experiment:latest" #container image for model training application
timestamp = datetime.now()
aip.init(project=PROJECT_ID,
         staging_bucket=STAGING_BUCKET,
         location=LOCATION)

disk_spec = {
    "boot_disk_type": "pd-ssd"  ,
    "boot_disk_size_gb": 100
}

machine_spec = {
    "machine_type": "n1-highmem-16",
    "accelerator_type": "NVIDIA_TESLA_T4",
    "accelerator_count": 1
}


containerSpec = {
    "image_uri": CONTAINER_URI,
    "args": [
        "--train_start=2023-05-01",
        "--train_end=2023-10-31",
        "--test_start=2023-11-01",
        "--test_end=2023-11-30",
        "--NUM_EPOCHS=150",
        "--VALIDATION_SPLIT=0.1",
        "--data_bucket=ficc-pipelines-test",
        "--file=similar_trades_df.pkl",
        "--BATCH_SIZE=10000",
        "--DROPOUT=0.2",
        "--LEARNING_RATE=0.0007",
        "--custom_model_path=gs://ficc-model-experiments/similar_trade_history_5yr_2LSTM_attention",
        f"--experiment_name=similar_trade_history_5yr_2LSTM_attention{timestamp.strftime('%Y-%d-%m-%H:%M')}"
    ]
}

worker_pool_spec = [
    {
        "replica_count": 1,
        "machine_spec": machine_spec,
        "disk_spec": disk_spec,
        "container_spec": containerSpec
    }
]
```

Create a custom job with arguments

```python
job = aip.CustomJob(display_name=JOB_NAME, 
                worker_pool_specs=worker_pool_spec)
```

```python
#the number of runs is a discrete parameter; linear refers to scaling for the hyperparameter and is a required argument, but here it is unimportant

n_runs = 5
runs_param = hpt.DiscreteParameterSpec(list(range(1,n_runs+1)), 'linear') 
```

```python
runs_param.values
```

Instantiate HyperparameterTuningJob class and pass arguments, then run on correct service account

```python
MAX_GPUS = 30 

hpt_job = aip.HyperparameterTuningJob(
    display_name=JOB_NAME,
    custom_job=job,
    metric_spec={
        "mae": "minimize",
    },
    parameter_spec={
        "experiment_number": runs_param,
    },
    max_trial_count=n_runs, 
    parallel_trial_count=min(MAX_GPUS, n_runs), #change if required; limited by GCP quotas on max GPUs used in training instances
)
```

```python
hpt_job.run(service_account = SERVICE_ACCOUNT)
```
